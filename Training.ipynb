```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

data = pd.read_csv("players/2544/data.csv")

data.head()
```

```python
data = data.drop(["GAME_ID", "PERIOD","LOC_X", "LOC_Y", "TEAM_ID", "TIME_PER","HME_AWA"], axis=1)
data.head()
```

```python
test = data.sample(frac=0.3)
train = pd.concat([data,test]).drop_duplicates(keep=False)

features = train.drop(["SHOT_MADE_FLAG"], axis=1)
labels = train["SHOT_MADE_FLAG"]

test_feat = test.drop(["SHOT_MADE_FLAG"], axis=1)
test_labels = test["SHOT_MADE_FLAG"]
```

```python
clf = MultinomialNB()
clf.fit(features, labels)

pred = clf.predict(test_feat)
accuracy = accuracy_score(test_labels, pred)
print(accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(100)
clf.fit(features, labels)

pred = clf.predict(test_feat)
accuracy = accuracy_score(test_labels, pred)
print(accuracy)
```

```python
from sklearn.svm import SVC

clf = SVC(kernel="linear")
clf.fit(features,labels)

pred = clf.predict(test_feat)
accuracy = accuracy_score(test_labels, pred)
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=20)
clf.fit(features,labels)

pred = clf.predict(test_feat)
accuracy = accuracy_score(test_labels, pred)
print(accuracy)
```
